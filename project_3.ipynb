# Проект III: Интерактивная карта на свободную тему

В этом задании вам предстоит создать **интерактивную веб-карту** на интересующую вас тему и опубликовать её онлайн.

В рамках задания мы:

- выберем тему и подберём к ней пространственные данные;
- подготовим и оформим несколько тематических слоёв;
- настроим интерактивную веб-карту (Folium) с разными способами изображения данных;
- добавим элементы управления и интерактивные инструменты;
- опубликуем карту и опишем её в README.


### Требования к итоговому результату

- **Слои:**
  - минимум **3 разных тематических слоя**;
  - минимум **2 разных способа изображения данных**, например:
    - картограмма (choropleth);
    - модель плотности (heatmap);
    - кластеризация маркеров.

- **Другие обязательные элементы:**
  - **Layer Control** (управление слоями);
  - минимум **2 дополнительных интерактивных инструмента** на выбор:
    - Mouse Position;
    - Fullscreen;
    - Mini Map;
    - Search;
    - или другие плагины Folium.

- **Публикация:**
  Карта должна быть размещена онлайн (рекомендуется `gh-pages` на GitHub, но можно использовать другие доступные способы).

- **README:**
  К карте прикладывается файл `README.md`, в котором вы даёте ссылку на опубликованную карту, кратко в свободной форме описываете, о чём ваша работа, и перечисляете исходные данные с указанием источников.


### Критерии оценивания (max. 10 баллов)

1. Слои и способы визуализации (4 балла)

- 2 балла – на карте присутствует минимум 4 разных тематических слоя.
- 2 балла – использовано минимум 3 разных способа изображения данных.

2. Интерактивность (2 балла)

- 1 балл – добавлен Layer Control (управление слоями).
- 1 балл – добавлено минимум 2 дополнительных интерактивных инструмента.

3. Публикация (2 балла)

- 2 балла – карта опубликована онлайн и открывается по рабочей ссылке.

4. Оформление итогового результата (2 балла)

- 1 балл – приложен README со ссылкой на карту, описанием работы и источниками данных.
- 1 балл – ноутбук структурирован, читаем и аккуратно оформлен с помощью Markdown.


### Как выполнять задание

В отличие от предыдущих проектов, **в этот раз мы не расписываем шаги по отдельности**. Тему, исходные данные, набор слоёв, способы визуализации и структуру ноутбука вы определяете **самостоятельно** — всё, что вы делаете, остаётся на ваш выбор.

Основной инструмент этого проекта — библиотека **Folium**. Перед началом работы **ознакомьтесь с её документацией** и посмотрите, какой функционал она предлагает: типы слоёв, маркеры, картограммы, плагины. Это поможет выбрать способы изображения данных и интерактивные инструменты для вашей карты.

> 🗺️ Документация Folium: [python-visualization.github.io/folium](https://python-visualization.github.io/folium/)

Именно потому, что шаги не расписаны, особенно важно **оформление самого ноутбука**. Сопровождайте код заголовками и пояснениями в Markdown-ячейках: что за данные, откуда они, что вы делаете на каждом этапе и почему. Аккуратно оформленная и читаемая работа оценивается выше.

> 📝 Памятка по Markdown, которой удобно пользоваться при оформлении ноутбука: [Markdown Guide — Cheat Sheet](https://www.markdownguide.org/cheat-sheet/)

> 🌐 Инструкция по публикации карты на GitHub Pages — в отдельном ноутбуке: [Публикация на GitHub Pages](../module_7/map_3.ipynb)


### Формат сдачи

Работа сдаётся в виде **ссылки на публичный GitHub-репозиторий**, который должен содержать:

- **Jupyter Notebook** с кодом + используемые данные;
- файл **`README.md`** с описанием (см. требования выше) и ссылкой на опубликованную карту;
- файл карты **`index.html`**, опубликованный через **GitHub Pages (`gh-pages`)**.

Карта должна быть доступна онлайн по ссылке из `README.md`.
